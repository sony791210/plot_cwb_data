```python
def convert_grid(data_nc,minlon,maxlon,minlat,maxlat,r,time):
    X0,Y0=np.arange(minlon,maxlon+r/2,r),np.arange(minlat,maxlat+r/2,r)
    XX,YY=np.meshgrid(X0,Y0)
    lon=data_nc['lon'][:].data
    lat=data_nc['lat'][:].data
    points = np.array( (lon.flatten(), lat.flatten()) ).T
    u_wind = data_nc['uwind'][:].data[time,:,:].flatten()
    v_wind = data_nc['vwind'][:].data[time,:,:].flatten()
    uu = griddata( points, u_wind, (XX,YY) , method='linear')
    vv = griddata( points, v_wind, (XX,YY) , method='linear')
    spd=(uu**2+vv**2)**0.5
    return uu,vv,spd,XX,YY
    
    
    
    
```

```python
u_wrf,v_wrf,spd_wrd,XX,YY=convert_grid(netCDF4.Dataset('sflux_air_1.003.nc'),116,128,16,30,0.1,12)
u_gfs,v_gfs,spd_gfs,XX,YY=convert_grid(netCDF4.Dataset('sflux_air_1.015.nc'),116,128,16,30,0.1,12)
```

```python
def go_plot(time):
    u_wrf,v_wrf,spd_wrd,XX,YY=convert_grid(netCDF4.Dataset('sflux_air_1.003.nc'),116,128,16,30,0.1,time)
    u_gfs,v_gfs,spd_gfs,XX,YY=convert_grid(netCDF4.Dataset('sflux_air_1.015.nc'),116,128,16,30,0.1,time)
    m=Basemap(projection='merc',llcrnrlat=16,urcrnrlat=30,\
                llcrnrlon=116,urcrnrlon=128,resolution='l') #resolution='h'
    xx,yy=m(XX,YY)
    fig = plt.figure()
    fig.set_size_inches(10, 6, forward=True)
    diff=spd_wrd-spd_gfs
    cs=m.contourf(xx,yy,diff,np.linspace(-5,5,51,endpoint=True),extend='both',cmap=cmo.balance)
    cbar = m.colorbar(cs, location='right', pad="5%")
#set fontsize in color bar 
    cbar.ax.tick_params(labelsize=20)
    m.drawcoastlines()
#m.fillcontinents(color='#AAAAAA',lake_color='aqua')
#title
    if (time==12):
        plt.title('2019/02/22 00:00'  ,fontsize=15,weight='bold')
    elif time==36:
        plt.title('2019/02/23 00:00'  ,fontsize=15,weight='bold')
    elif time==60:
        plt.title('2019/02/24 00:00'  ,fontsize=15,weight='bold')
    elif time==84:
        plt.title('2019/02/25 00:00'  ,fontsize=15,weight='bold')
    plt.xlabel('Longitude',fontsize=20, labelpad=35)
    plt.ylabel('Latitude',fontsize=20,  labelpad=35)
    fig.savefig(str(time)+'.png')
```

```python
go_plot(12)
go_plot(36)
go_plot(60)
go_plot(84)
```
