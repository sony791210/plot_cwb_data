```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
from matplotlib.font_manager import FontProperties
```

```python
# read data
data=pd.read_csv('particle.track.lonlat', sep='\s+',header=None)
```

```python
data.head()
```

```python
# define   point=5 runday=20   
# plot jump 20 plot once
points=5
days=20
jump=24
```

```python
#max_lon=data[0].max()+0.1
#min_lon=data[0].min()-0.1
#max_lat=data[1].max()+0.1
#min_lat=data[1].min()-0.1

# plot include taiwan
max_lon=123
min_lon=113
max_lat=27
min_lat=18
```

```python
# define lon lat
lon=data[0]
lat=data[1]

xlon=lon
ylat=lat

row_5=len(data)
row=row_5/points

lon_data=pd.DataFrame()
lat_data=pd.DataFrame()

# get 5 point lon lat
for i in range(points):
    lon_data[i]=xlon[0+24*i*days:24*days*(i+1)].reset_index(drop=True)
    lat_data[i]=ylat[0+24*i*days:24*days*(i+1)].reset_index(drop=True)
```

```python
m=Basemap(projection='merc',llcrnrlat=min_lat,urcrnrlat=max_lat,\
            llcrnrlon=min_lon,urcrnrlon=max_lon,resolution='l') #resolution='h'
```

```python
fig = plt.figure()
fig.set_size_inches(12, 8, forward=True)
#ax = plt.gca()
```

```python
# plot all line
xg,yg=m(lon_data[0].values,lat_data[0].values)
m.plot(xg,yg,'g')
```

```python
# plot jump point
fig = plt.figure()
fig.set_size_inches(12, 8, forward=True)
ct=0
st=1
out_x=[]
out_y=[]
for i,j in zip(lon_data[0].values,lat_data[0].values):
    if(ct==0):  
        xp,yp=m(i,j)
        m.plot(xp,yp,'r', marker='D')
        plt.text(xp, yp, str(st), fontsize=12);
        st=st+1
        out_x.append(i)
        out_y.append(j)
    ct=ct+1
    if(ct>=jump):
        ct=0
out_x.append(i)
out_y.append(j)
xp,yp=m(i,j)
m.plot(xp,yp,'r', marker='D')
plt.text(xp, yp, str(st), fontsize=12);
```

```python
# add title 
fig = plt.figure()
fig.set_size_inches(12, 8, forward=True)    
#make ylabel
m.drawparallels(np.arange(min_lat,max_lat+1,2),labels=[1,0,0,0],fontsize=15, linewidth=0.0)
#make xlabel
m.drawmeridians(np.arange(min_lon,max_lon+1,2),labels=[0,0,0,1],fontsize=15, linewidth=0.0)
m.drawcoastlines()

plt.xlabel('Longitude',fontsize=15, labelpad=30)
plt.ylabel('Latitude',fontsize=15 ,  labelpad=100)
plt.title('track  ',fontsize=15,weight='bold')
```

```python
# add table
fig = plt.figure()
fig.set_size_inches(12, 8, forward=True)
table=[]
st=1
for i,j in zip(out_x,out_y):
    i="{:.6f}".format(i)
    j="{:.6f}".format(j)
    table.append([st,i,j])
    st=st+1
table=np.array(table) 
rows=['num','lon','lat']

# control bbox=(0,-1.1,1,1) to adjust the table
the_table = plt.table(cellText=table,cellLoc='center',
                      colLabels=rows,bbox=(0,-1.1,1,1))

for (row, col), cell in the_table.get_celld().items():
    if (row == 0):
        cell.set_text_props(fontproperties=FontProperties(weight='bold'))
the_table.set_fontsize(14)
```
