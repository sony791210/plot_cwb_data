```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#read data   ,  the data is from u when mk vgrid.in
z_m=pd.read_csv('vgrid_master.out',sep='\s+',header=None)
z1=pd.read_csv('transect1.out',sep='\s+',header=None)
```

```python
np_m=z_m.shape[0]
nv_m=z_m.shape[1]-3
zcor_m=z_m.iloc[:,3:]
```

```python
for i in range(np_m):
    kbp_m=z_m.iloc[i,1]
    zcor_m.iloc[i,kbp_m:]=np.nan
```

```python
np_number=z1.shape[0]
nvrt=z1.shape[1]-6
zcor1=z1.iloc[:,6:]
```

```python
fig=plt.figure(figsize=(10, 6))
plt.plot(z1.iloc[:,4],zcor1,'k-',z1.iloc[:,4],-z1.iloc[:,5],'r.')
for i in range(np_number):
    plt.plot(z1.iloc[i,4]*np.ones((nvrt,1)),zcor1.iloc[i,:],'k')
plt.title('Transect before adjustment (transect1)');
plt.xlabel('Along transect distance (m)');
```

```python
final=os.path.join('vgrid.png')
fig.savefig(final)
```
