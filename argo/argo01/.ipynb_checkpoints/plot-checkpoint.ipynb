# INFO 為有出現在台灣周圍的argo表
# data.final為一段時間的內的ARGO 資料

```python
import os
os.environ['PROJ_LIB'] = '/opt/conda/share/proj'
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
```

# 定義顏色表　並讀入資料

```python

color=['#666666','#C10066','#E63F00','#DDAA00','#99DD00','#00DD77','#00DDDD','#0000CC','#7700BB','#990099','#EEEE00','#AAFFEE','#CCDDFF','	#5500FF']
argo=pd.read_csv('data.final',sep='\s+',header=None)
```

```python
argo_data=np.vstack(np.array([str(argo[0][i]).split('/')]) for i in range(len(argo)))
argo_list=argo_data[:,7].tolist()
a2=set(argo_list)
argo_cor={}
for i,j in zip(a2,color):
    argo_cor[str(i)]=j
    
argo_color=pd.Series(argo_list).map(argo_cor)        
```

```python
data=pd.read_csv('info',sep='\s+',header=None)
```

```python
argo_cc=[]
argo_name=[]
for i in data[0]:
    argo_cc.append(argo_color[i-1])
    argo_name.append(argo_list[i-1])
lon=data[2]
lat=data[4]
```

```python
fig = plt.figure()
m=Basemap(projection='merc',llcrnrlat=18,urcrnrlat=30,\
            llcrnrlon=116,urcrnrlon=128) #resolution='h'

fig.set_size_inches(10, 8, forward=True)
ax = plt.gca()
x,y=m(lon.tolist(),lat.tolist())

for i in range(len(x)):
    ax.scatter(x[i],y[i],color=argo_cc[i],marker='*',s=60,label=argo_name[i])

hand, labl = ax.get_legend_handles_labels()
handout=[]
lablout=[]
for h,l in zip(hand,labl):
   if l not in lablout:
        lablout.append(l)
        handout.append(h)
ax.legend(handout, lablout)
#plt.tight_layout()



m.readshapefile('shape/ocm2','comarques',linewidth=1.5,drawbounds=True,color='k')    
    
#draw shape
patches   = []
for info, shape in zip(m.comarques_info, m.comarques):
    patches.append(Polygon(np.array(shape), True) )
ax.add_collection(PatchCollection(patches, facecolor= '#AAAAAA', edgecolor='k', linewidths=1., zorder=2))

####################       label   ################################################  
      
#make ylabel
m.drawparallels(np.arange(18,31,2),labels=[1,0,0,0],fontsize=20, linewidth=0.0)
#make xlabel
m.drawmeridians(np.arange(116,130,4),labels=[0,0,0,1],fontsize=20, linewidth=0.0)
#title
plt.title('Oct.  Argo ' ,fontsize=15,weight='bold')  
        
plt.xlabel('Longitude',fontsize=20, labelpad=35)
plt.ylabel('Latitude',fontsize=20,  labelpad=55)
fig.savefig('argo.png')
```
